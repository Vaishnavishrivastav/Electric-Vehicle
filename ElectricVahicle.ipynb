```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=df, x='YEAR', y='TOTAL')
plt.title('Total EV Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

df_sorted = df.sort_values('YEAR')

sns.barplot(data=df_sorted, x='YEAR', y='TOTAL', color='skyblue')

plt.title('Total EV Sales by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

xticks = df_sorted['YEAR'].tolist()
plt.xticks(ticks=range(0, len(xticks), 5), labels=xticks[::5], rotation=45)

plt.tight_layout()
plt.show()
```

```python
ev_type_sales = df[['2 W', '3 W', '4 W', 'BUS']].sum().sort_values(ascending=False)
ev_type_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by EV Type')
plt.ylabel('Total Units Sold')
plt.xlabel('EV Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
from google.colab import files

# Plot and save the graph
ev_type_sales = df[['2 W', '3 W', '4 W', 'BUS']].sum().sort_values(ascending=False)
ev_type_sales.plot(kind='bar', color='lightgreen')

plt.title('Total Sales by EV Type')
plt.ylabel('Total Units Sold')
plt.xlabel('EV Type')

# Save the plot before showing it
plt.savefig('ev_type_sales.png')  # Save as PNG

plt.show()  # Show the plot

# Download the image file
files.download('ev_type_sales.png')  # Triggers download
```
